```python
# Program to check prime number or not
#include <stdio.h>
int main()
{
    int n,i,c=0;
    scanf("%d",&n);
    for(i=1;i<=n;i++)
    {
        if(n%i==0)
        c++;
    }
    if(c==2)
    printf("Prime number");
    else
    printf("Not");
    return 0;
}
```

```python
# Another method for prime number
#include <stdio.h>
int main()
{
    int n,i,c=0;
    scanf("%d",&n);
    for(i=2;i<=n/2;i++)
    {
        if(n%i==0)
        c++;
    }
    if(c==0)
    printf("Prime number");
    else
    printf("Not");
    return 0;
}
```

```python
# Finding nth fibonacci(only given n value) and count
#include <stdio.h>
int main()
{
    int n,f1=0,f2=1,f3,i,c;
    scanf("%d",&n);
    if(n==1)
    printf("%d",f1);
    else
    {
    for(i=3;i<=n;i++)
    {
       f3=f1+f2;
       f1=f2;
       f2=f3;
    }
    c=printf("%d",f2);
}
printf("\n%d",c);
}
```

```python
# Pattern(90 degree triangle with *)
#include <stdio.h>
int main()
{
    int i,j;
    for(i=1;i<=5;i++)
    {
    for(j=1;j<=i;j++)
    {
    printf("%d",i);
    if(j<i)
    printf("*");
    }
    printf("\n");
}
}
```

```python
# Pattern(equilateral triangle)
#include <stdio.h>
int main()
{
    int i,j,k;
    for(i=1;i<=5;i++)
    {
        for(k=5;k>=i;k--)
        {
            printf(" ");
        }
    for(j=1;j<=i;j++)
    {
    printf("%d ",j);
    }
    printf("\n");
}
}
```
