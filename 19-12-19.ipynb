```python
# Reverse of a number**
#include <stdio.h>

int main()
{
    int n,r;
    scanf("%d",&n);
    while(n!=0)
    {
        r=n%10;
        printf("%d",r);
        n=n/10;
    }
    return 0;
}
```

```python
# Sum of digits of a number
#include <stdio.h>
int main()
{
    int n,r,s=0;
    scanf("%d",&n);
    while(n!=0)
    {
        r=n%10;
        s=s+r;
        n=n/10;
    }
    printf("%d",s);
    return 0;
}
```

```python
# Sum and count of the number
#include <stdio.h>
int main()
{
    int n,r,s=0,c=0;
    scanf("%d",&n);
    while(n!=0)
    {
        c++;
        r=n%10;
        s=s+r;
        n=n/10;
    }
    printf("%d %d",s,c);
    return 0;
}
```

```python
# Reverse of a number
#include <stdio.h>
int main()
{
    int n,r,s=0;
    scanf("%d",&n);
    while(n!=0)
    {
        r=n%10;
        s=s*10+r;
        n=n/10;
    }
    printf("%d",s);
    return 0;
}
```

```python
# Palindrome number or not
#include <stdio.h>
int main()
{
    int t,n,r,s=0;
    scanf("%d",&n);
    t=n;
    while(t!=0)
    {
        r=t%10;
        s=s*10+r;
        t=t/10;
    }
    if(n==s)
    printf("Palindrome number");
    else
    printf("Not");
    return 0;
}
```

```python
# Armstrong number or not
#include <stdio.h>
int main()
{
    int t,n,r,s=0;
    scanf("%d",&n);
    t=n;
    while(t!=0)
    {
        r=t%10;
        s=s+r*r*r;
        t=t/10;
    }
    if(n==s)
    printf("Armstrong number");
    else
    printf("Not");
    return 0;
}
```

```python
# Armstrong or not for n numbers
#include <stdio.h>
#include <math.h>
int main()
{
    int n,r,t,s=0,c=0;
    scanf("%d",&n);
    t=n;
    while(t!=0)
    {
        c++;
        t=t/10;
    }
    t=n;
    while(t!=0)
    {
        r=t%10;
        s=s+pow(r,c);
        t=t/10;
    }
    if(n==s)
    printf("Armstrong number");
    else
    printf("not");
    return 0;
}
```

```python
# Armstrong range
#include <stdio.h>
#include <math.h>
int main()
{
    int n,r,t,s=0,c=0,range;
    scanf("%d",&range);
    for(n=1;n<=range;n++)
    {
    c=0;
    s=0;
    t=n;
    while(t!=0)
    {
        c++;
        t=t/10;
    }
    t=n;
    while(t!=0)
    {
        r=t%10;
        s=s+ceil(pow(r,c));
        t=t/10;
    }
    if(n==s)
    printf(" %d",n);
    }
}
```
